# folium.Marker 

```python
import pandas as pd
library = pd.read_excel('병원정보_최신화.xlsx')
library.head()
```

```python
library.shape
```

```python
import folium
m = folium.Map(location=[35.196735, 128.625025], tiles='openstreetmap', zoom_start=10)
```

```python
from folium import Marker
for _, row in library.iterrows():
    Marker(location = [row['YPos'], row['XPos']],
           popup=row['addr'],
           icon=folium.Icon(color='blue',icon='star')
          ).add_to(m)
m
```

# folium.MarkerCluster

```python
from folium.plugins import MarkerCluster

# Draw a basemap
m = folium.Map(location=[35.196735, 128.625025], tiles='openstreetmap', zoom_start=11)

# Add  points to the map
mc = MarkerCluster()
for _, row in library.iterrows():
    mc.add_child(    
        Marker(location = [row['YPos'], row['XPos']],
               popup=row['addr']
              )
    )
    
m.add_child(mc)


# Display the map
m
```

# Heatmap 시각화 

```python
import pandas as pd
library = pd.read_excel('병원정보_최신화.xlsx')
import folium
```

```python
from folium.plugins import HeatMap

# Draw a basemap
m = folium.Map(location=[35.196735, 128.625025], tiles='openstreetmap', zoom_start=11)

# Add heatmap
HeatMap(
    data=library[['YPos', 'XPos']], 
    radius=20,
).add_to(m)

# Display the map
m
```

# Choropleth Map 시각화 

```python
!pip install geopandas
import geopandas as gpd
library = pd.read_csv('./Library.csv')
seoul_boundary = gpd.read_file('./서울시구경계.shp', encoding='cp949')
```

```python
seoul_boundary.rename(columns={'SIGUNGU_NM':'구명'}, inplace=True)
```

```python
seoul_boundary.rename(columns={'SIGUNGU_NM':'구명'}, inplace=True)
boundary = seoul_boundary[['구명', 'geometry']].set_index('구명')
```

```python
library_count = library.구명.value_counts()
```

```python
print(seoul_library.crs)     # 첫번째 결과 epsg:4326
print(seoul_boundary.crs)    # 두번째 결과 epsg:5186 
```

```python
ax = seoul_boundary.plot(figsize=(10,10), color='whitesmoke', edgecolor='black', linewidth=1)
ax.axis('off')
seoul_library.to_crs(epsg=5186).plot(ax=ax, color='deepskyblue', markersize=10)
```

```python
# Create a base map
m = folium.Map(location=[37.58, 127.0], tiles='cartodbpositron', zoom_start=11)

# Add a choropleth map
Choropleth(
    geo_data=boundary['geometry'].geometry.to_crs(epsg=4326).__geo_interface__,
    data=library_count,
    key_on='feature.id',
    fill_color='YlGnBu',
    legend_name='Seoul Libraries',
).add_to(m)

# Display the map
m
```
