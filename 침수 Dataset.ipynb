### 침수 지역별 일일 1시간 최다 강수량 max

```python
import pandas as pd
df_nc1 = pd.read_csv("북면_강수_201408_201408.csv", encoding = 'CP949')
df_nc1.sort_values('날짜', ascending=True)
```

```python
# [3] 20060711 경상남도 창원시 의창구 대산면
df_nc2 = df_nc1.groupby('일자').agg(Maximum_precipitation_per_hour = (' 강수량', 'max')).iloc[1:, :]

df_nc2.sort_values('Maximum_precipitation_per_hour', ascending=False)

#df_nc2.sort_values(['한시간_최대_강수량', '날짜'], ascending=[False, False])
#df2.sort_values('일자', ascending=False)
```

####  표1

```python
df_nc2['Maximum_precipitation_per_hour'].plot.hist()
```

#### 표2 

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정
plt.rcParams['font.family'] = 'NanumGothic'

plt.figure(figsize=(12, 30))
sns.countplot(y=df_nc2['Maximum_precipitation_per_hour'], order=df_nc2['Maximum_precipitation_per_hour'].value_counts().index)
plt.yticks(fontsize=12)
plt.title('the_most_precipitation_in_an_hour_per_day')
plt.show()
```
