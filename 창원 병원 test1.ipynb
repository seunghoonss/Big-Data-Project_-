```python
import pandas as pd
```

```python
df = pd.read_csv("병원정보_2022.csv")
df
```

```python
df_nc1 = df.query('sgguCdNm == ["창원마산회원구" , "창원마산합포구", "창원성산구", "창원진해구", "창원의창구"]' or 'XPos>=1' or 'YPos>=1').sort_values(['XPos', 'YPos'], ascending=[False, False]) 


df_nc2 = df_nc1.iloc[:, 3:29].drop(columns = ['emdongNm', 'postNo', 'telno', 'hospUrl', 'estbDd',
                                   'drTotCnt', 'mdeptGdrCnt', 'mdeptIntnCnt', 'mdeptResdntCnt', 'mdeptSdrCnt',
                                   'detyGdrCnt', 'detyIntnCnt', 'detyResdntCnt', 'detySdrCnt', 'cmdcGdrCnt',
                                   'cmdcIntnCnt', 'cmdcResdntCnt', 'cmdcSdrCnt'])

df_nc2.loc[~df_nc2["XPos"].isnull()] # Nan이 아닌값만 추출
```

# 각 (5개) 구에 대한 창원 병원 현황  

```python
# 파생변수 만들기
# 집단별로 요약하는 Qroupby
# 추출한 데이터를 이용해 창원구(5개)에 대한 집단별로 요약

df_nc2.groupby('sgguCdNm').agg(ct_XPos = ('XPos', 'count'),
                              ct_YPos = ('YPos', 'count')).sort_values(['ct_XPos', 'ct_YPos'], ascending=[False, False]) 
```

# 병원 종류(class)에 대한 창원 병원 현황

```python
# 파생변수 만들기
# 집단별로 요약하는 Qroupby
# 추출한 데이터를 이용해 병원종류(class)에 대한 집단별로 요약

df_nc2.groupby('clCdNm').agg(ct_XPos = ('XPos', 'count'),
                              ct_YPos = ('YPos', 'count')).sort_values(['ct_XPos', 'ct_YPos'], ascending=[False, False]) 
```
