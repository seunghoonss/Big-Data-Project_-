```python
# 해당 장소를 클릭했을 때 위도 경도 반환하는 LaLngPopup() 함수
import folium 

m = folium.Map(
    location =[37.544129, 127.054357],
    zoom_start = 14,
    tiles = 'OpenStreetMap') 

m.add_child(folium.ClickForMarker(popup = 'ClickFormMarker')) # folium 라이브러리에는 지도 위에 마우스를 클릭했을 때 마커를 표시해주는 함수
m.add_child(folium.LatLngPopup())

folium.CircleMarker([37.544129, 127.054357], radius=100, popup='코엑스 범위').add_to(m)
m
```
